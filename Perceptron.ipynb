```python
import numpy as np
def step(z):
    if z > 0:
        return 1
    else:
        return 0
```

```python
def perceptron(x,w,b):
    z = np.dot(x,w) + b
    y = step(z)
    return y
```

```python
w1 = np.random.rand(1)
w2 = np.random.rand(1)
w = np.array([w1,w2])
b = np.random.rand(1)
y = 1
lr = 0.01
```

```python
def update_weights(y2,y,w1,w2,b):
    error = y2-y
    w1 = w1-error*lr
    w2 = w2-error*lr
    b = b-error*lr
    return w1,w2,b
```

```python
test1 = np.array([0, 1])
test2 = np.array([1, 1])
test3 = np.array([0, 0])
test4 = np.array([1, 0])
```

```python
epochs = 50
for i in range (0,epochs):
    y2 = perceptron(test1,w,b)
    w1,w2,b = update_weights(y2,y,w1,w2,b)
    print("w1 = ",w1,"w2 = ",w2,"b = ",b)
```
